The numbers generated by random.random are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.

Generate 1000 numbers from random.random and plot their PMF and CDF. Is the distribution uniform?

```python
import random
randomNumbers = []
for i in range(1000):
    randomNumbers.append(random.random())

# Create PMF
import thinkstats2
randomNumbersPmf = thinkstats2.Pmf(randomNumbers, label='random.random')
```

```python
%matplotlib inline
import thinkplot
import numpy as np
thinkplot.PrePlot(1)
thinkplot.Pmf(randomNumbersPmf)
thinkplot.Show(xlabel='random number between 0 and 1', ylabel='PMF', yticks=np.arange(0.0,0.0014,0.0002))
```

Judgeing from the probability mass function (PMF) graph, random.random() indeed generates a uniform distribution of random numbers.

Now let's plot the cumulative distribution function:

```python
randomNumbersCdf = thinkstats2.Cdf(randomNumbers, label='random.random')
thinkplot.PrePlot(1)
thinkplot.Cdf(randomNumbersCdf)
thinkplot.Show(xlabel='random number between 0 and 1', ylabel='CDF')
```

The uniformity of the distribution is confirmed by the almost perfectly straight slope of the CDF function.
