### Exercise 7.1

Using data from the NSFG, make a scatter plot of birth weight versus mother’s age. Plot percentiles of birth weight versus mother’s age. Compute Pearson’s and Spearman’s correlations. How would you characterize the relationship between these variables?

```python
%matplotlib inline
import nsfg
import thinkplot
import thinkstats2

df = nsfg.ReadFemPreg()
df = df.dropna(subset=['agepreg','totalwgt_lb'])

sample = thinkstats2.SampleRows(df, 6000)
birthweights_lbs, mothers_age_years = sample.totalwgt_lb, sample.agepreg

thinkplot.Scatter(birthweights_lbs, mothers_age_years, alpha=0.08)
thinkplot.Show(xlabel='Birth weight (lbs)',
               ylabel="Mother's age (years)",
               axis=[0,15,10,50])
```

We see a strong vertical separation right at the 6 pound value, possibly due to rounding bias.

```python
# Plot the percentiles
import numpy as np

bins = np.arange(0,15,1)
indices = np.digitize(df.totalwgt_lb, bins)
groups = df.groupby(indices)

for i, group in groups:

    birth_weights = [group.totalwgt_lb.mean() for i, group in groups]
    cdfs = [thinkstats2.Cdf(group.agepreg) for i, group in groups]
    
    for percent in [75,50,25]:
        ages = [cdf.Percentile(percent) for cdf in cdfs]
        label = '%dth' % percent
        thinkplot.Plot(birth_weights,
                       ages,
                       label=label)
```

Between 3 and 11 pounds, the flatness of the graphs suggest the absence of a (linear) relationship between both variables. The correlation statistics confirm this insight:

```python
print "Pearson's correlation: %1.4f" % thinkstats2.Corr(birthweights_lbs, mothers_age_years)
```

```python
print "Spearman's rank correlation: %1.4f" % thinkstats2.SpearmanCorr(birthweights_lbs, mothers_age_years)
```

Both values are close to zero, so we can safely conclude that there is hardly any (linear) relationship between birth weight and the age of the mother.
